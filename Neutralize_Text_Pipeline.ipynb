## Prepare Data
Chunk by paragraphs in to dictionary with space for neutralized text.

```python
import re

def chunk_text(raw_text, title):
  raw_text = re.sub(r'\n\s*\n', '\n\n', raw_text)  # Normalize paragraph breaks
  raw_text = re.sub(r'[ \t]+', ' ', raw_text)       # Normalize spaces
  raw_chunks = raw_text.split('\n\n')
  cleaned_chunks = []

  for i, chunk in enumerate(raw_chunks):
    if len(chunk) > 25:
      d = {}
      d['id'] = i + 1
      d['author'] = author_name
      d['target'] = chunk
      d['input'] = ""
      d['title'] = title
      cleaned_chunks.append(d)

  return cleaned_chunks

moby_chunks = chunk_text(moby_text, "Moby Dick")
bartleby_chunks = chunk_text(bartleby_text, "Bartleby")
billy_chunks = chunk_text(billy_text, "Billy Budd")

# sample structure
for chunk in moby_chunks[:3]:
  print('Id: \n', chunk['id'],'\n')
  print('Title: ', chunk['title'])
  print('Author: ','\n', chunk['author'],'\n')
  print('Target: ','\n', chunk['target'],'\n')
  print('Input: ', '\n', chunk['input'],'\n')
```
